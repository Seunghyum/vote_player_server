```python
import json
from collections import Counter
from pathlib import Path
import pandas as pd

# 1. íŠ¹ì • í´ë” ê²½ë¡œ ì„¤ì •
folder_path = Path('/Users/seunghyunmoon/Code/Toy/vote_player_server/data/candidates')  # <-- ì—¬ê¸°ë¥¼ ë³¸ì¸ í´ë”ë¡œ ì„¤ì •í•˜ì„¸ìš”

# ê²°ê³¼ ê¸°ë¡ìš© ë¦¬ìŠ¤íŠ¸
summary_records = []
empty_bills_files = []  # bills ì—†ëŠ” íŒŒì¼ ëª©ë¡

# 2. í´ë” ë‚´ ëª¨ë“  JSON íŒŒì¼ ìˆœíšŒ
json_files = list(folder_path.glob('*.json'))

print(f"ì´ {len(json_files)}ê°œì˜ JSON íŒŒì¼ì„ ì²˜ë¦¬í•©ë‹ˆë‹¤.\n")

for file_path in json_files:
    print(f"ì²˜ë¦¬ ì¤‘: {file_path.name}")

    # 2-1. íŒŒì¼ ì½ê¸°
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    # 2-2. ê¸°ì¡´ billsStatusStatisticsê°€ ìˆìœ¼ë©´ ì‚­ì œ
    if 'billsStatusStatistics' in data:
        print(f"  - ê¸°ì¡´ billsStatusStatistics ì»¬ëŸ¼ì„ ì‚­ì œí•©ë‹ˆë‹¤.")
        del data['billsStatusStatistics']

    # 2-3. bills ë°ì´í„° ìˆ˜ì§‘
    bills = data.get('bills', [])
    bills_count = len(bills)

    # ğŸ‘‰ billsê°€ ì•„ì˜ˆ ì—†ëŠ” ê²½ìš° ê²½ê³  ì¶œë ¥ + ê¸°ë¡
    if bills_count == 0:
        print(f"  âš ï¸  ê²½ê³ : ì´ íŒŒì¼ì€ bills ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤!")
        empty_bills_files.append(file_path.name)

    counter = Counter()
    for bill in bills:
        nth = (bill.get('nth') or '').replace('ì œ', '').replace('ëŒ€', '')
        status = (bill.get('status') or '').strip()
        if nth and status:
            counter[(nth, status)] += 1

    # 2-4. Counterë¥¼ ë¦¬ìŠ¤íŠ¸ë¡œ ë³€í™˜
    bills_status_statistics = [
        {
            'nth': f"{nth}ëŒ€",
            'name': status,
            'value': count
        }
        for (nth, status), count in counter.items()
    ]

    statistics_count = len(bills_status_statistics)

    # 2-5. ê²°ê³¼ë¥¼ ë°ì´í„°ì— ì¶”ê°€
    data['billsStatusStatistics'] = bills_status_statistics

    # 2-6. ê²°ê³¼ íŒŒì¼ ì €ì¥ (ì›ë˜ íŒŒì¼ ë®ì–´ì“°ê¸°)
    with open(file_path, 'w', encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

    print(f"  - ì €ì¥ ì™„ë£Œ: {file_path.name}\n")

    # 2-7. ìš”ì•½ ê¸°ë¡ ì €ì¥
    summary_records.append({
        'íŒŒì¼ëª…': file_path.name,
        'bills í•­ëª© ìˆ˜': bills_count,
        'í†µê³„ í•­ëª© ìˆ˜': statistics_count
    })

# 3. ìš”ì•½ í…Œì´ë¸” ì¶œë ¥
summary_df = pd.DataFrame(summary_records)

# ğŸ‘‰ 'í†µê³„ í•­ëª© ìˆ˜' ê¸°ì¤€ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
summary_df = summary_df.sort_values(by='í†µê³„ í•­ëª© ìˆ˜', ascending=False)

print("âœ… ì²˜ë¦¬ ê²°ê³¼ ìš”ì•½:")
display(summary_df)

# 4. billsê°€ ì—†ëŠ” íŒŒì¼ ëª©ë¡ ì¶œë ¥
if empty_bills_files:
    print("\nâš ï¸ bills ë°ì´í„°ê°€ ì—†ëŠ” íŒŒì¼ ëª©ë¡:")
    for filename in empty_bills_files:
        print(f" - {filename}")
else:
    print("\nâœ… ëª¨ë“  íŒŒì¼ì— bills ë°ì´í„°ê°€ ì¡´ì¬í•©ë‹ˆë‹¤!")
```
